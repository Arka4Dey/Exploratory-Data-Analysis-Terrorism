```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
url = "/content/globalterrorismdb_0718dist.csv"
data = pd.read_csv(url,encoding="latin1")
print("Data imported successfully")

data.head()
```

**All Columns**

```python
data.columns.values
```

```python
terror_data = data[['iyear','imonth','iday','country_txt','region_txt','provstate','city','attacktype1_txt','gname','weaptype1_txt','targtype1_txt','weapdetail','nkill','nwound','summary']]
```

```python
terror_data.head()
```

```python
terror_data.isnull().sum()
```

## **Country With Most Terror Activities**

```python
terror_data['country_txt'].value_counts().head()
```

```python
plt.figure(figsize=(30,12))
sns.countplot('country_txt', data=terror_data)
plt.title('Country With Most Terrorist activities',fontsize=30)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(terror_data['country_txt'].value_counts()[:10].index, terror_data['country_txt'].value_counts()[:10].values)
plt.show()
```

### **These are the top 10 Countries with Most Terror activities. Iraq is at the top**

## **Terrorist attack Per Year**

```python
plt.figure(figsize=(30,12))
sns.countplot('iyear', data=terror_data)
plt.xlabel("Year",fontsize=25)
plt.ylabel("Attack Count",fontsize=25)
plt.title('Terrorist Attack Per Year',fontsize=40)
plt.xticks(rotation=90)
plt.show()
```

### **Above Graph shows us, in year 2014 most terrorist attack was carried out.**

## **People Killed in Terror Attack**

```python

df = terror_data[['iyear','nkill']].groupby('iyear').sum()
fig,ax4 = plt.subplots(figsize=(20,6))
df.plot(kind='bar',alpha=0.7,ax=ax4)
plt.title("People Killed Due To Attack",fontsize=25)
plt.ylabel("Number of killed peope",fontsize=20)
plt.xlabel('Year',fontsize=20)
top_side = ax4.spines["top"]
top_side.set_visible(False)
right_side = ax4.spines["right"]
right_side.set_visible(False)
```

### **From Above the graph most people were killed in 2014.**

## **Weapons Used**

```python
plt.figure(figsize=(15,6))
sns.countplot('weaptype1_txt', data=terror_data)
plt.xlabel("Weapons",fontsize=15)
plt.ylabel("Attack Count",fontsize=15)
plt.title('Weapons Used',fontsize=30)
plt.xticks(rotation=90)
plt.show()
```

### **Explosive was the most common weapons used in terror attacks.**

## **Region with Maximum Attack**

```python
plt.figure(figsize=(15,6))
sns.countplot('region_txt', data=terror_data)
plt.xlabel("Place",fontsize=15)
plt.ylabel("Attack Count",fontsize=15)
plt.title('Region with Maximum Attack',fontsize=25)
plt.xticks(rotation=90)
plt.show()
```

### **Middle East and North Africa Region were most of Attack took Place.**

## **Most Target Place**


```python
plt.figure(figsize=(15,6))
sns.countplot(terror_data['targtype1_txt'], order = terror_data['targtype1_txt'].value_counts().index)
plt.xlabel("Region",fontsize=15)
plt.ylabel("Attack Count",fontsize=15)
plt.title('Target Site',fontsize=25)
plt.xticks(rotation=90)
plt.show()
```

### **Private Citizens and Property are most target Site of Terrorist**

## **Organisation Involved In Attacks**

```python

plt.figure(figsize=(8,6))
sns.barplot(terror_data['gname'].value_counts()[:15].values, terror_data['gname'].value_counts()[:15].index)
plt.xlabel("Attack Count",fontsize=15)
plt.ylabel("Groups",fontsize=15)
plt.title('Terror Organisation',fontsize=30)
plt.xticks(rotation=90)
plt.show()
```

### **Terror Organisation with most Terror attacks is Taliban**

## **Hot Zone of Terrorism**

```python
plt.figure(figsize=(15,6))
sns.barplot(terror_data['city'].value_counts()[:15].index, terror_data['city'].value_counts()[:15].values)
plt.xlabel("City",fontsize=15)
plt.ylabel("Attack Count",fontsize=15)
plt.title('Hot zone of terrorism.',fontsize=35)
plt.xticks(rotation=90)
plt.show()
```

### **From above graph Baghdad is the hot zone of terrorism**
